Merge Joining and concatination

```python
students = {
    "Student_ID": [101, 102, 103, 104],
    "Name": ["Ali", "Sara", "Ahmed", "Zoya"],
    "City": ["Karachi", "Lahore", "Islamabad", "Karachi"]
}
marks = {
    "Student_ID": [101, 102, 104, 105],
    "Marks": [85, 92, 88, 75],
    "Grade": ["A", "A+", "B+", "B"]
} 
df_students=pd.DataFrame(students)
df_students

```

```python

df_marks=pd.DataFrame(marks)
df_marks
```

```python
pd.merge(df_students, df_marks, on='Student_ID')
```

```python
pd.merge(df_students, df_marks, on='Student_ID',how='outer')
```

```python
pd.merge(df_students, df_marks, on='Student_ID',how='right')
```

Concatination

```python


df1 = pd.DataFrame({
    "Name": ["Ali", "Sara"],
    "Marks": [85, 90]
})

df2 = pd.DataFrame({
    "Name": ["Ahmed", "Zoya"],
    "Marks": [75, 88]
})

pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], axis=1)
```

joining

```python
df1 = pd.DataFrame({
    "Name": ["Ali", "Sara", "Ahmed"],
    "Marks": [85, 90, 78]
}, index=[1, 2, 3])

df2 = pd.DataFrame({
    "City": ["Karachi", "Lahore", "Islamabad"]
}, index=[1, 2, 3)
```

```python
 df1.join(df2)

```

Group By 

```python
data = {
    "Department": ["IT", "HR", "IT", "HR"],
    "Salary": [50000, 40000, 60000, 45000]
}
df=pd.DataFrame(data)
df
```

```python
df.groupby('Salary')['Department'].sum()
```

Aggregation

```python
df['Salary'].mean()
```

```python
df.groupby("Department")["Salary"].agg(
    ["mean","median", "max", "min"]
)
```

Pivot Table

```python

data = {
    "Name": ["Ali", "Sara", "Ahmed", "Zoya"],
    "Department": ["IT", "HR", "IT", "HR"],
    "Salary": [50000, 60000, 55000, 65000]
}

df = pd.DataFrame(data)
df
```

```python
pd.pivot_table(  df,  values="Salary",  index="Department", aggfunc="mean"
)
```

```python
pd.crosstab(df["Department"], df["Name"])
```
